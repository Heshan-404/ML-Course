```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error,accuracy_score
```

```python
data = pd.read_csv("../CSV/GPA_data.CSV")
data.head()
```

```python
dummy = pd.get_dummies(data[["Gender","Extra_Curricular"]], dtype=int)
dummy
```

```python
dummy.drop(['Gender_Female','Extra_Curricular_No'], axis=1,inplace=True)
dummy
```

```python
data.drop(['Gender','Extra_Curricular','ID'],axis=1,inplace=True)
#here id should also removed because ID is an unnecessary data coloumn
data
```

```python
data_new = pd.concat([dummy,data], axis=1)
data_new
```

```python
x = data_new.iloc[:, :-1]
y = data_new.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```
