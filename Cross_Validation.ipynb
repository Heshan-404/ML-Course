```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv("./CSV/Boston.csv")
data.head()
```

```python
x = data.iloc[:,:12]
y = data.iloc[:,12]
```

```python
model = LinearRegression()
model
```

```python
cvals = -cross_val_score(model,x,y,cv = 5 ,scoring='mean_squared_error')
cvals #this is not valid 
```

```python
cvals = -cross_val_score(model,x,y,cv = 5 ,scoring='neg_mean_squared_error')
cvals
#scikit learn prefers higher scores are better.
#MSE is an error metric where lower is better
#This gives a higher value of negative value. That mean negative form of negative error.
```

```python
RMSE = np.sqrt(cvals)
RMSE
```

```python
np.mean(RMSE)
```

```python
cvals_r2 = cross_val_score(model,x,y,cv = 5 ,scoring='r2')
cvals_r2  
```

```python
np.mean(cvals_r2)
```
